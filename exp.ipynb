```python
df = df[df['type'] != 'Quadricycle']
```

```python

sns.barplot(x=df['value'],y=df['type'])
```

* In this graph we can clearly see that Two-Wheelers segment dominate the Indian market sales from 2007 to 2019 and this is likely will continue .
* In the Two-Wheelers , the most selling type of two-wheelers is motor cycles after that scooters sells the most which is nearly less than  half of the sales of the Motor cycle.
* After the Two-Wheelers the segment which has most sales is Passenger cars which include all the types of Passenger cars like Hatchback, Sedan, SUV, MUV, Coupe, Pickup Trucks etc.
* But the sales unit is less as compare to the Two-Wheelers .
* After that commercial vehicles sales the most in India.

```python
sns.barplot(x=df['Year'],y=df['value'])
```

* In this graph, we can see that the yearly sales is increasing linearly from 2007-1018.
* The sales of the 2019 is slightly decreased because of the COVID-19.
* After COVID-19 the sales of the vehicle is likely to be grow linearly as previous decade.

```python
plt.figure(figsize=(8,8))
sns.barplot(x=df['Year'],y=df['value'],hue=df['type'])
plt.show()
```

* This is a graph of types of vehicles sales every year
* In this graph we can see that Two-Wheelers has increased significantly every year and slow down in 2019 because of covid.
* But the motor cycle increased from 2007 to 2014 but stayed constant after that while scooters was not selling well from 2007 to 2013 but after that it start selling exponentially and this is likely will be constant like this.
* While passenger car stayed almost constant throughout the year from 2007 to 2019 
* commercial vehicles also not increased as compared to others .
